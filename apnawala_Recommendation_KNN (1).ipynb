```python
import pandas as pd
from pathlib import Path
import json
```

```python

# path to file
os = Path(r'apnawala.json')


# read json
with os.open('r', encoding='utf-8') as f:
    data = json.loads(f.read())
```

```python
# Normalizing data
#Normalize semi-structured JSON data into a flat table.

df1 = pd.json_normalize(data, record_path =['items'])
df1
```

```python
#Extracting important features from json file

df1 = df1[['id','snippet.title']]
df1
```

```python
#Rename the common feature from both the dataframes
df = df.rename(columns={"category_id": "id"})
df
```

```python
#converting the feature datatype int
df['id']=df['id'].astype(int)
df1['id']=df1['id'].astype(int)
```

```python
# to merge both the dataframes on common feature
df_merge = df1.merge(df, on='id')
df_merge.head(20)
```

```python
#total likes for each channel

rating = df.dropna(axis = 0, subset = ['channel_title'])
ratingcount = (rating.groupby(by = ['channel_title'])['likes'].
          count().
          reset_index().
          rename(columns = {'likes': 'totallikesCount'})
          [['channel_title', 'totallikesCount']])
ratingcount.head()
```

```python
# merge the dfs
likes_with_totallikesCount = rating.merge(ratingcount, left_on = 'channel_title', right_on = 'channel_title', how = 'left')
likes_with_totallikesCount.head()
```

```python
#describe total counts 
# pd.set_option('display.float_format', lambda x: '%.3f' % x)
# print(ratingcount['totallikesCount'].describe())
```

```python
#histogram for total likes 
plt.figure(figsize=(10,4))
ratingcount['totallikesCount'].hist(bins=30)
```

```python
# querying the total count based on threshold
threshold = 1
rating_popular_channel= likes_with_totallikesCount.query('totallikesCount >= @threshold')
rating_popular_channel.head()
```

```python

rating_popular_channel.shape
```

```python
## First lets create a Pivot matrix

channel_features_df=rating_popular_channel.pivot_table(index='channel_title',columns='id',values='likes').fillna(0)
channel_features_df.head()
```

```python
channel_features_df.to_csv('data.csv')
```

```python
# KNN with hyper parameter tuning
#cosine
#brute

import pickle
from scipy.sparse import csr_matrix

channel_features_df_matrix = csr_matrix(channel_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(channel_features_df_matrix)

with open('knn_model', 'wb') as file:
    pickle.dump(model_knn, file)
```

```python
channel_features_df.shape
```

```python
#Randomly selecting a row 

query_index = 3
# np.random.choice(channel_features_df.shape[0)
print(query_index)
distances, indices = model_knn.kneighbors(channel_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```
